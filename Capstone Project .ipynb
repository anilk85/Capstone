```python
Sno_num = []
Movie_title = []
Director_Name = []
Movie_Duration = []
Movie_year = []
Movie_ratings = [] 
Movie_Metascore = []
```

```python
# Scraping Serial number ...

num_tags = driver.find_elements(By.XPATH, "//span[@class='lister-item-index unbold text-primary']")
num_tags[0:4]
```

```python
num_tags
```

```python
# Now the text of the Sno is inside the tags extracted above ...

# so will be run a loop to iterate over the tags extracted above and extract the text inside there..

for i in num_tags:
    
    num = i.text
    Sno_num.append(num)
    
Sno_num[0:50]
```

```python
# Scraping the Movie title ...

title_tags = driver.find_elements(By.XPATH, "//h3[@class='lister-item-header']")
title_tags[0:4]
```

```python
title_tags
```

```python
# Now, extract the duration inside the tags, so will be run a loop to iterate over the tags extracted inside there...

for i in title_tags:
    
    title = i.text
    Movie_title.append(title)
    
Movie_title[0:50]
```

```python
# Scraping Movies year ...

year_tags = driver.find_elements(By.XPATH, "//span[@class='lister-item-year text-muted unbold']")
year_tags[0:4]
```

```python
# Now, extract text of the year inside the tags, so will be run a loop to iterate over the tags extracted inside there..

for i in year_tags:
    
    year = i.text
    Movie_year.append(year)
    
Movie_year[0:50]
```

```python
Duration_tags = driver.find_elements(By.XPATH, "//span[@class='runtime']")
Duration_tags[0:4]
```

```python
Duration_tags
```

```python
# Now, extract the duration inside the tags, so will be run a loop to iterate over the tags extracted inside there...

for i in Duration_tags:
    
    Duration = i.text
    Movie_Duration.append(Duration)
    
Movie_Duration[0:50]
```

```python
# Scraping the rating ...

ratings_tags = driver.find_elements(By.XPATH, "//div[@class='inline-block ratings-imdb-rating']")
ratings_tags[0:4]
```

```python
ratings_tags
```

```python
# Now, extract the duration inside the tags, so will be run a loop to iterate over the tags extracted inside there...

for i in ratings_tags:
    
    ratings = i.text
    Movie_ratings.append(ratings)
    
Movie_ratings[0:50]
```

```python
# Scraping the Metascore...

Metascore_tags = driver.find_elements(By.XPATH, "//div[@class='ratings-bar']")
Metascore_tags[0:4]
```

```python
Metascore_tags
```

```python
# Now, extract the duration inside the tags, so will be run a loop to iterate over the tags extracted inside there...

for i in Metascore_tags:
    
    Metascore = i.text
    Movie_Metascore.append(Metascore)
    
Movie_Metascore[0:50]
```
